```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
X = np.loadtxt("x.txt")
Y_d = np.loadtxt("out1.txt")
Y_sq = np.loadtxt("out2.txt")
Y_ex = np.loadtxt("out3.txt")
```

```python
fig  = plt.figure(figsize = (8, 4))
ax = plt.subplot(111)

a = ["double runge", "square runge", "precise"]
ax.plot(X, Y_d,'g', X, Y_sq, 'b^', X, Y_ex, 'r--')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend(a)
ax.grid()
```
